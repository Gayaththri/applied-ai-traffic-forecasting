# Background & Literature Review

In urban places across the world, traffic congestion is a serious problem that results in longer travel times, higher fuel use, pollution of the environment, and a lower standard of living. Accurate and timely information about future traffic conditions is necessary for successful traffic control as cities expand and transportation networks become more complicated. Therefore, a key element of Intelligent Transportation Systems (ITS) is short-term traffic flow and congestion prediction, which enables proactive traffic management, dynamic routing, and informed decision-making by drivers and city planners. Artificial Intelligence (AI) has emerged as a crucial enabler in resolving these issues due to the explosive growth of traffic data from loop detectors, GPS-enabled automobiles, and connected infrastructure.

Traditional statistical and machine learning methods including linear regression, support vector regression (SVR), and autoregressive integrated moving average (ARIMA) models were the mainstays of early traffic forecasting methodologies (Awad et al., 2015; Jin et al., 2022). These techniques offered interpretable, inexpensive baselines and showed how past traffic patterns could be used to forecast near-future situations. These models, however, are ill-suited to representing the intricate, nonlinear temporal dynamics and spatial interconnections that define major urban traffic networks because they presuppose linear linkages and stationarity. As a result, when congestion circumstances change quickly, their prediction performance drastically deteriorates.

Neural network-based techniques, specifically multilayer perceptrons (MLPs), were developed to get around these restrictions. MLPs can include a variety of input variables, including past traffic flow, road characteristics, and outside variables like weather, and they can represent nonlinear connections (Liu et al., 2024). Although MLPs are more flexible than classical models, they do not explicitly reflect the spatial relationships between connected road segments and still interpret traffic data primarily as independent time series. In traffic control scenarios, where congestion frequently spreads across nearby roadways rather than happening in isolation, this restriction is crucial.

Architectures that jointly describe the temporal and spatial properties of traffic data have been the focus of recent developments in deep learning. A popular method is the hybrid CNN-LSTM architecture, which combines long short-term memory (LSTM) networks for temporal modeling with convolutional neural networks (CNNs) for spatial feature extraction (Zhang et al., 2022; Luo et al., 2024). While LSTM layers simulate temporal dependencies and traffic evolution over time, CNN layers in this architecture learn spatial correlations across various traffic sensors, successfully capturing congestion patterns across road segments. For short-term traffic prediction, CNN-LSTM models considerably outperform conventional machine learning and standalone neural network models, especially for horizons up to 60 minutes ahead, according to empirical research utilizing benchmark datasets like METR-LA and PEMS-BAY. For instance, Zhang et al. (2022) demonstrated that CNN-LSTM models are capable of precisely capturing congestion waves moving across urban roadways.

Concurrently, graph-based deep learning techniques have been created to clearly depict traffic networks' non-Euclidean structure. By modeling traffic sensors as nodes connected by spatial or functional linkages, Graph Convolutional Networks (GCNs) and Spatio-Temporal Graph Neural Networks (STGNNs) enable more accurate depiction of network architecture (Yu et al., 2018; Hong et al., 2022). In order to extract long-term trends and periodic patterns from lengthy historical data, more modern models, such the Long-Short Term Transformer Network (LSTTN), use transformer designs and masked sub-series learning (Luo et al., 2024). In a similar vein, TITAN presents a heterogeneous mixture-of-experts paradigm that integrates sequence-centric, variable-centric, and prior-knowledge-centric modeling, outperforming earlier state-of-the-art methods by up to 11% (Wang et al., 2024).

Transformer-based and graph-centric models perform well, but they frequently need a lot of processing power, big training datasets, and intricate model architecture. This may restrict their applicability for real-time implementation in city-scale traffic management systems, especially in situations where resilience, computational economy, and interpretability are crucial. CNN-LSTM models, on the other hand, provide a favorable trade-off between deployment practicalities, model complexity, and predictive accuracy. They work particularly well for short-term forecasting tasks, when it is more crucial to capture current traffic patterns and geographical correlations than to represent very long-term trends (Li et al., 2023; Xu et al., 2021).

AI methods are being utilized more and more in traffic control applications, including demand forecasting, adaptive signal timing, incident detection, and route optimization, in addition to congestion prediction (Chen et al., 2021; Jin et al., 2024). By minimizing stop-and-go traffic, traffic authorities can reduce delays, improve travel time reliability, and reduce emissions through proactive control strategy adjustments made possible by accurate short-term forecasts. As decision-support tools in intelligent transportation systems, short-horizon, data-driven models like CNN-LSTM are essential.

In conclusion, artificial intelligence (AI) applications in traffic control have progressed from traditional statistical techniques to neural networks and sophisticated deep learning architectures that incorporate temporal and spatial modeling. The current state of the art is represented by graph-based and transformer-based models, although CNN-LSTM architectures offer a good compromise between computational economy, performance, and interpretability. CNN-LSTM models are an adequate and justified choice for the this prototype since they are especially well adapted for short-term urban traffic congestion prediction.

# Model Selection & Comparison

### Goal

This research aims to use historical traffic flow data to forecast short-term traffic congestion levels on a metropolitan road network one hour ahead of time. In addition to making precise forecasts, the goal is to assess which Artificial Intelligence (AI) methods are appropriate for traffic control applications. Prediction accuracy, data needs, computing complexity, setup and result time, and practical viability for real-world intelligent transportation systems are all taken into account in this assessment.

Three AI methods are chosen and contrasted: Regression, Multilayer Perceptron (MLP), and CNN-LSTM. These methods represent progressively higher levels of model complexity and provide a structured basis for comparison.

### Technique 1: Regression

One of the oldest and most used techniques for predicting traffic flow is regression-based. Because of their ease of use and interpretability, Support Vector Regression (SVR) and Linear Regression are frequently used as baseline models in traffic forecasting literature. These techniques assume a reasonably constant relationship between past and future traffic circumstances and model traffic flow as a function of previous observations.

Strengths:
Regression models can be trained fast even on big datasets, are simple to implement, and need little processing power. Because of their great interpretability, traffic engineers can comprehend how specific characteristics like past flow or speed affect forecasts. Regression approaches are widely utilized as benchmarking tools in traffic forecast studies because of these characteristics.

Weaknesses:
Regression techniques' incapacity to represent non-linear temporal dependencies and spatial correlations across road networks is its main drawback. Urban traffic patterns are dynamic by nature and are impacted by intricate relationships between various route segments. Regression models have trouble capturing peak-hour dynamics and congestion propagation because they treat inputs independently.

Data & Output:
Tabular time-series data, such as past traffic volume or speed observations, are usually needed for these models. One numerical prediction (such as traffic flow or speed at a future time step) is the output.

Setup Time and Computational Cost:
Regression models train very quickly, even on big datasets, and require very little preprocessing. Because predictions are generated nearly instantly, they can be used for quick testing and real-time baseline deployment.

Suitability:
Regression offers a solid baseline against which more sophisticated AI methods may be assessed, even though it is inappropriate for high-accuracy traffic management systems.

### Technique 2: Neural Networks

By adding hidden layers and nonlinear activation functions, Multilayer Perceptrons (MLPs) expand regression techniques and enable the modeling of more intricate traffic patterns. When there are nonlinear correlations between past traffic measurements and predicted congestion levels, MLPs are frequently used for short-term traffic prediction jobs.

Strengths:
Regression models are unable to capture non-linear temporal patterns, however MLPs can. They are adaptable, relatively simple to implement, and a natural step up from linear regression. MLPs work well for prototyping since they demand less compute and train faster than deep spatio-temporal models.

Weaknesses:
MLPs do not explicitly capture spatial relationships between road segments and treat inputs as flat vectors, despite their capacity to represent non-linearities. Because congestion on one route directly affects nearby roads, this reduces their usefulness for network-wide traffic forecast. Furthermore, if MLPs are not properly regularized, they may overfit.

Data & Output:
Structured numerical input, usually flattened historical time-series data from several sensors, is needed for MLPs. A forecasted traffic value for a future time span is the output.

Setup Time and Computational Cost:
Compared to regression models, MLPs take a significant amount of setup and tweaking work. On normal hardware, the training period is greater but still reasonable, and once the model is trained, predictions are produced rapidly.

Suitability:
MLPs are an intermediate technique rather than a state-of-the-art solution since they provide better performance than regression but are still limited for complex urban traffic networks.

### Technique 3: Deep Learning (CNN-LSTM)

CNN-LSTM models integrate Long Short-Term Memory (LSTM) networks for temporal dependency modeling with Convolutional Neural Networks (CNNs) for spatial feature extraction. Many people consider this hybrid architecture to be among the best methods for predicting traffic flow and congestion in the short term.

Strengths:
Both temporal changes over time and spatial correlations between road segments are explicitly captured by CNN-LSTM models. While LSTM layers simulate traffic evolution and peak-hour trends, CNN layers learn patterns of spatial congestion. Research utilizing datasets like METR-LA and PeMS shows that CNN-LSTM performs noticeably better in prediction accuracy than regression and MLP models. The architecture closely matches the needs of traffic control in the actual world.

Weaknesses:
Larger datasets, longer training times, and more processing power are needed for CNN-LSTM models. Additionally, they are harder to understand than more straightforward models, which can be problematic in explainable AI scenarios.

Data & Output:
Typically, input data is multi-dimensional time-series data organized to depict road network spatial layouts. Each road segment's short-term traffic flow or congestion forecast is the result.

Setup Time and Computational Cost:
CNN-LSTM models take a long time to set up, including training, architectural design, and data preprocessing. Training requires a lot of processing power and frequently benefits from GPU acceleration. But once taught, inference is quick enough to anticipate traffic in real time.

Justification for Selection:
Because CNN-LSTM provides the best balance between predictive performance and practical applicability for smart city traffic control, it is chosen for the main implementation. Its capacity to simulate spatiotemporal connections makes it a good fit for short-term congestion forecasting in urban settings, despite being more complicated than regression and MLP models.

Summary Comparison

In conclusion, CNN-LSTM offers the most thorough representation of spatiotemporal traffic dynamics at the expense of increased computational complexity, MLPs enhance non-linear modeling but overlook spatial structure, and regression models offer quick and comprehensible baselines but lack expressive power. The choice of CNN-LSTM as the main method for this prototype is supported by these trade-offs.

## b)

Since traffic flow prediction is essentially a temporal forecasting problem, the system makes use of multivariate time-series traffic data. The main dataset chosen for this research is the METR-LA Traffic Dataset, which includes actual traffic measurements gathered from 207 loop detectors placed along Los Angeles' main thoroughfares. Over 30,000 timestamps are obtained by recording the data at 5-minute intervals over several months, meeting the need for a substantial data volume appropriate for deep learning and machine learning models. The METR-LA dataset is a trustworthy and reputable baseline for assessing traffic forecasting models since it is publicly accessible and frequently utilized in traffic prediction research.

The dataset is organized as a multivariate time series in its raw form, with each row denoting a timestamp and each column corresponding to observations of traffic flow or speed from a particular sensor. A sliding window technique is used to modify the data in order to enable supervised learning using a CNN–LSTM model. In particular, future traffic conditions are predicted using traffic data from the preceding 12 time steps, which is comparable to one hour of historical data (12 × 5 minutes). The input data takes on a three-dimensional structure of (samples, time_steps, features) following this transformation. A processed dataset, for instance, might be shaped like (10,000, 12, 207), with each sample representing a one-hour traffic sequence across all sensors. This format works well for LSTM layers to model temporal dependencies and CNN layers to capture local spatial patterns across sensors.

To guarantee stable model training and high-quality data, a number of preparation procedures are used. To maintain temporal continuity, forward filling or mean interpolation techniques are used to address missing values, which frequently result from transient sensor failures or communication problems. In order to prevent features with bigger magnitudes from controlling the learning process and to improve convergence during CNN–LSTM training, Min–Max normalization is then used to scale all traffic values into the range [0,1]. In order to maintain chronological order and prevent data leaking, the dataset is then divided into 70% training data and 30% testing data using a time-aware split without shuffling. For efficient short-term traffic congestion prediction, these preprocessing procedures guarantee that the incoming data is clear, consistent, and properly formatted.

## c)

```python
# Import Required Libraries
import os
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, LSTM, Dense, Dropout

# Plot configuration for clarity
plt.style.use("seaborn-v0_8")
plt.rcParams["figure.figsize"] = (10, 4)
```

```python
# Dataset Download
dataset_url = "https://zenodo.org/record/5146275/files/METR-LA.csv?download=1"
dataset_filename = "METR-LA.csv"

if not os.path.exists(dataset_filename):
    print("Downloading METR‑LA dataset from Zenodo…")
    response = requests.get(dataset_url)
    if response.status_code == 200:
        with open(dataset_filename, "wb") as f:
            f.write(response.content)
        print("✓ Download complete!")
    else:
        raise Exception(f"Failed to download dataset. Status code: {response.status_code}")
else:
    print("✓ Dataset already exists locally.")
```

```python
# Load and Inspect Data
data = pd.read_csv(dataset_filename)

print("Raw data shape:", data.shape)
print("Columns:", data.columns.tolist()[:10], "...")
display(data.head())
```

```python
# Remove timestamp/index column, keep sensor readings only
traffic_data = data.iloc[:, 1:].values
num_time_steps, num_sensors = traffic_data.shape
print(f"\nExtracted traffic_data shape: {traffic_data.shape} (time_steps, sensors)")
```

```python
# Data Preprocessing
# Handle Missing Values
traffic_df = pd.DataFrame(traffic_data)
traffic_df = traffic_df.fillna(method="ffill")   # forward fill
traffic_df = traffic_df.fillna(traffic_df.mean())  # remaining NaNs with column means
traffic_data = traffic_df.values
```

```python
# Normalisation
scaler = MinMaxScaler(feature_range=(0, 1))
traffic_scaled = scaler.fit_transform(traffic_data)
```

```python
# Sliding Window Creation
# Converts time-series data into supervised learning format
def create_sliding_windows(data_array, window_size=12):
    """
    data_array: (time_steps, num_sensors)
    returns:
        X: (samples, window_size, num_sensors)
        y: (samples, num_sensors)
    """
    X, y = [], []
    for i in range(len(data_array) - window_size):
        X.append(data_array[i:i + window_size])
        y.append(data_array[i + window_size])
    return np.array(X), np.array(y)

time_steps = 12  # 12*5 min = 1 hour history
X, y = create_sliding_windows(traffic_scaled, time_steps)

print("\nProcessed input shape X:", X.shape)
print("Processed output shape y:", y.shape)
```

```python
# Train-Test Split (Time-Aware)
# Chronological split prevents data leakage
split_index = int(0.7 * len(X))
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

print("\nTrain input shape:", X_train.shape)
print("Test input shape:", X_test.shape)
```

```python
# CNN-LSTM Model Architecture
model = Sequential()

# CNN layer learns spatial relationships between sensors
model.add(
    Conv1D(
        filters=64,
        kernel_size=3,
        activation="relu",
        input_shape=(X_train.shape[1], X_train.shape[2]),
    )
)
model.add(Dropout(0.2))

# LSTM layer learns temporal dependencies
model.add(LSTM(64, activation="tanh"))
model.add(Dropout(0.2))

# Output layer predicts next-step traffic speed for all sensors
model.add(Dense(X_train.shape[2], activation="linear"))

model.compile(optimizer="adam", loss="mse", metrics=["mae"])
model.summary()
```

```python
# Model Training
history = model.fit(
    X_train,
    y_train,
    epochs=15,
    batch_size=32,
    validation_split=0.1,
    verbose=1,
)
```

```python
# Model Evaluation
y_pred = model.predict(X_test)

# Metrics in normalised scale
mae_norm = mean_absolute_error(y_test, y_pred)
rmse_norm = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"\nTest MAE (normalized): {mae_norm:.4f}")
print(f"Test RMSE (normalized): {rmse_norm:.4f}")

# Convert back to real-world units (mph)
y_test_inv = scaler.inverse_transform(y_test)
y_pred_inv = scaler.inverse_transform(y_pred)

mae_real = mean_absolute_error(y_test_inv, y_pred_inv)
rmse_real = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))

print(f"Test MAE (mph, approx): {mae_real:.4f}")
print(f"Test RMSE (mph, approx): {rmse_real:.4f}")
```

```python
# Visualisation
# Training Loss
plt.figure()
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("CNN‑LSTM Training & Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Prediction vs Actual (Single Sensor)
plt.figure()
plt.plot(y_test[:100, 0], label="True (Sensor 0)")
plt.plot(y_pred[:100, 0], label="Predicted (Sensor 0)")
plt.title("CNN‑LSTM Traffic Prediction — Sensor 0 (100 samples, normalized)")
plt.xlabel("Time Step")
plt.ylabel("Normalized Speed")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

steps_slice = slice(0, 50)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.heatmap(y_test[steps_slice].T, cmap="RdYlGn_r")
plt.title("Actual Speeds (first 50 test steps)")
plt.xlabel("Time Step")
plt.ylabel("Sensor ID")

plt.subplot(1, 2, 2)
sns.heatmap(y_pred[steps_slice].T, cmap="RdYlGn_r")
plt.title("Predicted Speeds (first 50 test steps)")
plt.xlabel("Time Step")
plt.ylabel("Sensor ID")

plt.tight_layout()
plt.show()
```

```python
# Real-Time Prediction Demonstration (New Data)
# Simulates how the model behaves when new sensor data arrives

print("Simulating real-time prediction with NEW incoming data...")

latest_window = traffic_scaled[-time_steps:]
latest_window = np.expand_dims(latest_window, axis=0)

print("New input window shape:", latest_window.shape)

next_hour_prediction = model.predict(latest_window)

next_hour_prediction_real = scaler.inverse_transform(next_hour_prediction)

print("\nPredicted traffic speeds for NEXT HOUR (first 10 sensors):")
print(np.round(next_hour_prediction_real[0][:10], 2))
```

```python
# Visualize predicted traffic for next hour
plt.figure(figsize=(10, 4))
plt.plot(next_hour_prediction_real[0][:50])
plt.title("Predicted Traffic Speeds for Next Hour (First 50 Sensors)")
plt.xlabel("Sensor Index")
plt.ylabel("Speed (mph)")
plt.grid(True)
plt.show()
```

## Model Evaluation

### Testing Method and Baseline Comparison

Using the METR-LA test set, many assessment techniques are used to confirm the accuracy of the constructed CNNLSTM traffic forecasting model. Standard forecasting error measures are employed since traffic prediction is a regression-based time-series problem:

The average absolute difference between expected and actual traffic speeds is measured by the Mean Absolute Error (MAE).

Larger prediction errors are
penalized more severely by Root Mean Squared Error (RMSE), which is crucial during times of congestion when erroneous forecasts can be expensive.

**Baseline Models for Verification**

The CNN-LSTM's performance is evaluated against two baselines to make sure it learns significant patterns rather than pointless persistence:

1. Baseline for naive persistence

        y^t + 1 = yt


This baseline is predicated on the idea that traffic conditions won't change in the subsequent time step.

2. Baseline for Linear Regression:
A traditional machine learning baseline that ignores temporal and spatial structure was taught on flattened historical windows. Compared to naive persistence, this offers a more robust classical comparison.

On the same test set, both baselines must be surpassed by an accurate and successful AI implementation.





```python
# Naive persistence baseline: y(t+1) = y(t)
y_test_shifted = y_test[1:]
y_naive = y_test[:-1]

mae_naive_norm = mean_absolute_error(y_test_shifted, y_naive)
rmse_naive_norm = np.sqrt(mean_squared_error(y_test_shifted, y_naive))

print("BASELINE VS MODEL (NORMALIZED)")
print(f"Naive baseline  - MAE: {mae_naive_norm:.4f}, RMSE: {rmse_naive_norm:.4f}")
print(f"CNN-LSTM model  - MAE: {mae_norm:.4f},   RMSE: {rmse_norm:.4f}")
```

```python
# Sanity check: value ranges in real mph

print("\n SANITY CHECK: REAL-SCALE VALUES (mph)")
print("Sample of original test speeds (first 5 timesteps, first 5 sensors):")
print(np.round(y_test_inv[:5, :5], 2))

print("\nSample of predicted test speeds (first 5 timesteps, first 5 sensors):")
print(np.round(y_pred_inv[:5, :5], 2))
```

```python
# Per-sensor error analysis

sensor_mae = np.mean(np.abs(y_test_inv - y_pred_inv), axis=0)
best_sensor = int(np.argmin(sensor_mae))
worst_sensor = int(np.argmax(sensor_mae))

print("\nPER-SENSOR MAE (mph)")
print(f"Best sensor ID:   {best_sensor} with MAE ≈ {sensor_mae[best_sensor]:.2f} mph")
print(f"Worst sensor ID:  {worst_sensor} with MAE ≈ {sensor_mae[worst_sensor]:.2f} mph")
print(f"Average MAE over all sensors: {sensor_mae.mean():.2f} ± {sensor_mae.std():.2f} mph")
```

### Strengths and Limitations of the Implementation

**Strengths**

1. **Spatio-temporal modelling.**  

By simultaneously modeling temporal dynamics (by LSTM) and spatial correlations between sensors (by the convolutional layer), the CNN-LSTM represents the state of the art in traffic forecasting.

2. **Improvement over baseline.**  

The model clearly outperforms the naive "last value" baseline in both MAE and RMSE, indicating that it captures important predictive patterns rather than merely repeating the previous observation.

3. **Stable training behaviour.**  

The gradual convergence of training and validation losses without notable divergence indicates limited overfitting and good generalization to unobserved time periods.

4. **Rich visual evaluation.**  

Even if the internal weights are complicated, traffic engineers can understand the model's behavior thanks to heatmaps, per-sensor, and per-timestep error charts.

**Limitations**

1. **Black-box nature.**

CNN-LSTM does not explain individual forecasts. For critical infrastructure, decision-makers may use interpretable models or post-hoc explanation strategies like feature attribution or attention visualization.

2. **Sensitivity to rare events.**

Error spikes during odd times (such accidents or anomalous congestion peaks) show that the model underestimates rare events that are underrepresented in the training data.

3. **Limited feature set.**  

The implementation uses historical speed as its only input. It ignores exogenous variables like incidents, weather, holidays, or special events that have been shown to improve accuracy in more complex models.

4. **Computational cost.**  

Training a CNN-LSTM for multiple sensors and long histories is more expensive than using conventional baselines like ARIMA or SVR. Real-time large-scale deployment may require model reduction, pruning, or more efficient architectures to meet latency and energy constraints.

All things considered, the implemented CNN-LSTM shows excellent predictive performance, a noticeable improvement over a naive baseline, and provides useful insights for immediate traffic management while also pointing out areas for further development.

## Part E - Equity, Diversity, Inclusion (EDI) and Sustainability

### EDI: Bias, Fairness, Accessibility and Ethics

The METR-LA dataset, which includes traffic speed readings from 207 fixed sensors placed on Los Angeles' main thoroughfares, is used in this study. The sensor placement is concentrated on high-capacity, economically significant road routes rather than being fairly distributed throughout all neighborhoods, which causes structural bias even though the dataset is technically sound. Because of this, the CNN-LSTM model learns traffic patterns mostly for private car users on highways; smaller streets, public transportation routes, and residential or lower-class regions are underrepresented.

Concerns regarding algorithmic fairness in intelligent traffic management are raised by this uneven data coverage. The system would typically prioritize traffic flow on well-instrumented routes if predictions from this model were used to optimize ramp metering, route advice, or signal timings. Congestion and erratic travel times may persist in areas with inadequate sensing infrastructure, which could exacerbate already-existing socio-spatial disparities. These results are indicative of larger issues in smart-city research, where data-driven optimization may inadvertently help groups that are already privileged.

From the standpoint of accessibility and inclusion, the current implementation just considers vehicle speeds and ignores users of public transportation, bicycles, pedestrians, and people with disabilities. This restricts its capacity to facilitate inclusive transportation planning. Incorporating multi-modal data and adopting equity-aware goals, such as distributing the benefits of congestion reduction among various districts and modes of transportation rather than concentrating on private car users, would make a traffic intelligence system more equitable.

Compared to camera-based or trajectory-level systems, the use of aggregated speed data without personal identification lowers immediate privacy issues in terms of ethics and privacy. Transparency and accountability, however, continue to raise ethical questions. Predictive model-informed decisions, like reallocating road space or changing signal priorities, can have significant social repercussions. Therefore, ethical deployment would necessitate systems for human monitoring and public accountability, regular fairness audits across neighborhoods, and unambiguous documentation of system objectives.

Overall, even though the CNN-LSTM model has the potential to increase traffic efficiency, its implementation without fairness-aware design and governance may inadvertently exacerbate already-existing disparities in urban mobility.


### Environmental Sustainability: Compute, Storage and Green Deployment

The computational and energy needs of deep learning models like CNN-LSTM result in a non-trivial environmental footprint. Compared to traditional forecasting techniques like ARIMA or linear regression, training the model often requires GPU acceleration and entails recurrent optimization over huge spatiotemporal datasets. This particular experiment has a negligible environmental impact, but similar models that are regularly retrained and implemented at the municipal size could significantly increase carbon emissions.

When predictions need to be made continually for hundreds of sensors, computing costs are reduced but still substantial throughout inference time. Because of the continuous data transfer between sensors and data centers, centralized cloud deployment significantly raises energy usage. These elements emphasize how crucial it is to take sustainability into account both during long-term operational deployment and model training.

There are a number of ways to lessen these systems' negative effects on the environment. While maintaining the majority of predicted accuracy, model compression techniques like pruning, quantization, and knowledge distillation can drastically reduce model size and energy consumption. By making predictions closer to the point of data generation, edge or dispersed deployment can lower latency and communication overhead. Energy consumption associated with storage can also be decreased by using effective data storage practices, such as archiving older records in aggregated form and only keeping high-resolution data for current periods.

Lastly, accuracy and resource consumption must be balanced for sustainability. If a model significantly lowers energy consumption and operating expenses, it may be preferred in many real-world scenarios, even if it is slightly less sophisticated and little less accurate. AI-based traffic prediction systems can promote both better urban mobility and environmentally sustainable smart-city development by fusing precise forecasting with effective model design, deployment, and data management.
